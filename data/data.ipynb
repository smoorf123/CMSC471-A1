```python
import pandas as pd

df = pd.read_csv("Crimes_-_2001_to_Present_20260303.csv") # too big to store in github (~2.2gb)
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')
df = df.dropna(subset=['Date'])

crime_mapping = {
    'THEFT': 'THEFT',
    'BATTERY': 'BATTERY',
    'CRIMINAL DAMAGE': 'CRIMINAL_DAMAGE',
    'NARCOTICS': 'NARCOTICS',
    'BURGLARY': 'BURGLARY',
    'MOTOR VEHICLE THEFT': 'MOTOR_VEHICLE_THEFT'
}
df['Primary Type'] = df['Primary Type'].map(crime_mapping)
df_filtered = df.dropna(subset=['Primary Type']).copy()

df_filtered['Year'] = df_filtered['Date'].dt.year
df_filtered['Month_Num'] = df_filtered['Date'].dt.month      # 1 to 12
df_filtered['Day_Num'] = df_filtered['Date'].dt.dayofweek    # 0=Monday, 6=Sunday
df_filtered['Hour'] = df_filtered['Date'].dt.hour            # 0 to 23
```

```python
# heatmap dataset
heatmap_df = df_filtered.groupby(
    ['Primary Type', 'Day_Num', 'Hour']
).size().reset_index(name='IncidentCount')

heatmap_df = heatmap_df.sort_values(by=['Primary Type', 'Day_Num', 'Hour'])
heatmap_df.to_json('heatmap_data.json', orient='records')
```

```python
# line graph dataset
linegraph_df = df_filtered.groupby(
    ['Primary Type', 'Year', 'Month_Num']
).size().reset_index(name='IncidentCount')

linegraph_df = linegraph_df.sort_values(by=['Primary Type', 'Year', 'Month_Num'])
linegraph_df.to_json('linegraph_data.json', orient='records')
```
